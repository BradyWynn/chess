```python
import numpy as np
import matplotlib.pyplot as plt
from torch import nn
import torch
import sqlite3
import chess
import graphviz
import torch.autograd as autograd
```

```python
n = 10
m = 10
a = torch.rand((n, m))
```

```python
plt.imshow(a);
```

```python
p = 5
t = torch.randn((n, p), requires_grad=True)
w = torch.randn((p, m), requires_grad=True)
```

```python
loss_function = nn.MSELoss()
optimizer = torch.optim.AdamW([t, w], lr=0.001)
```

```python
for i in range(1000):
    optimizer.zero_grad()
    a_ = t @ w
    loss = loss_function(a_, a)
    print(loss)
    loss.backward()
    optimizer.step()
```
