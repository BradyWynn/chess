```python
def fen_to_bitboard(fen):
    bitboard = torch.zeros((12, 64), device=device)
    i = 0
    for letter in fen:
        if(letter == "/"):
            continue;
        try:
            i += int(letter)
        except:
            if(letter == "p"):
                bitboard[0, i] = 1
            if(letter == "n"):
                bitboard[1, i] = 1
            if(letter == "r"):
                bitboard[2, i] = 1
            if(letter == "b"):
                bitboard[3, i] = 1
            if(letter == "q"):
                bitboard[4, i] = 1
            if(letter == "k"):
                bitboard[5, i] = 1

            if(letter == "P"):
                bitboard[6, i] = 1
            if(letter == "N"):
                bitboard[7, i] = 1
            if(letter == "R"):
                bitboard[8, i] = 1
            if(letter == "B"):
                bitboard[9, i] = 1
            if(letter == "Q"):
                bitboard[10, i] = 1
            if(letter == "K"):
                bitboard[11, i] = 1
            i += 1
        if(i == 64):
            break
    return bitboard.view((1, 768))
```
